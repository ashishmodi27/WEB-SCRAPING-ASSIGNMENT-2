### Q4: Write a python program to scrape data for first 10 job results for Data scientist Designation in Noida location. You have to scrape company_name, No. of days ago when job was posted, Rating of the company.

```python
url = ' https://www.glassdoor.co.in/index.htm'
driver.get(url)
```

```python
#Login part is done manually becuase need to hide password
```

```python
from selenium.webdriver.common.keys import Keys
search_job = driver.find_element_by_id("sc.keyword") #Job title
search_job.send_keys("Data Scientist")
search_loc = driver.find_element_by_id("sc.location") #Location
search_loc.clear() #clearing the output
time.sleep(2)
search_loc.send_keys(Keys.CONTROL + "a")  #ctrl+a will select all stuff
time.sleep(2)
search_loc.send_keys(Keys.DELETE)  #clear selected stuff
search_loc.send_keys("Noida")

search_btn = driver.find_element_by_xpath("//button[@class='gd-ui-button ml-std col-auto SearchStyles__newSearchButton css-iixdfr']")
search_btn.click()

time.sleep(4)

company_names=[]
days=[]
urls=[]
ratings=[]

companies_tags=driver.find_elements_by_xpath("//a[@class=' css-l2wjgv e1n63ojh0 jobLink']")
for i in range(0,10):
    company_names.append(companies_tags[i].text)
    
post_day=driver.find_elements_by_xpath("//div[@class='d-flex align-items-end pl-std css-mi55ob']")
for i in range(0,10):
    days.append(post_day[i].text)
    
job_url=driver.find_elements_by_xpath("//a[@class='jobLink css-1rd3saf eigr9kq2']")
for i in range(0,10):
    urls.append(job_url[i].get_attribute('href'))
    
for i in urls:
    driver.get(i)
    try:
        rating_tags=driver.find_element_by_xpath("//span[@class='css-1pmc6te e11nt52q4']")
        ratings.append(rating_tags.text.replace("\n",' '))
    except NoSuchElementException as e:
        ratings.append("N/A")
        
jobs=pd.DataFrame({})
jobs['Company'] = company_names
jobs['Rating'] = ratings
jobs['Posting Date'] = days
jobs
```

### Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %

```python
url = 'https://www.flipkart.com/'
driver.get(url)

btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']") #closing login popup
btn.click()

search_job = driver.find_element_by_class_name("_3704LK") #search bar
search_job.send_keys("sunglasses")

search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']") #search button
search_btn.click()

time.sleep(4)

Brand=[]
Description=[]
Price=[]
Discount=[]
for i in range(0,3):
    brand_name=driver.find_elements_by_xpath("//div[@class='_2B099V']/div[1]")
    for i in brand_name:
        Brand.append(i.text)
    desc=driver.find_elements_by_xpath("//div[@class='_2B099V']/a[1]")
    for i in desc:
        Description.append(i.text)
    glass_price=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in glass_price:
        Price.append(i.text)
    prod_disc=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span")
    for i in prod_disc:
        Discount.append(i.text)
    
    nxt=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    url=[]
    for i in nxt:
        url=i.get_attribute('href')
    time.sleep(2)
    driver.get(url)
    
Glass=pd.DataFrame({})
Glass['Brand'] = Brand
Glass['Description'] = Description
Glass['Price'] = Price
Glass['Discount'] = Discount
Glass
```

### Q8: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field. You have to scrape 4 attributes of each sneaker :

```python
url = 'https://www.flipkart.com'
driver.get(url)

search_job = driver.find_element_by_class_name("_3704LK")
search_job.send_keys("sneakers")

search_btn = driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()

time.sleep(4)

Brand=[]
Description=[]
Price=[]
Discount=[]
for i in range(0,3):
    brand_name=driver.find_elements_by_xpath("//div[@class='_2B099V']/div[1]")
    for i in brand_name:
        Brand.append(i.text)
    desc=driver.find_elements_by_xpath("//div[@class='_2B099V']/a[1]")
    for i in desc:
        Description.append(i.text)
    Shoes_price=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in Shoes_price:
        Price.append(i.text)
    prod_disc=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span")
    for i in prod_disc:
        Discount.append(i.text)
    
    nxt=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    url=[]
    for i in nxt:
        url=i.get_attribute('href')
    time.sleep(2)
    driver.get(url)
    
Shoes=pd.DataFrame({})
Shoes['Brand'] = Brand[0:100]
Shoes['Description'] = Description[0:100]
Shoes['Price'] = Price[0:100]
Shoes['Discount'] = Discount[0:100]
Shoes
```

### Q9: Go to the link - https://www.myntra.com/shoes Set Price filter to “Rs. 6649 to Rs. 13099” , Color filter to “Black”, as shown in the below image.

```python
url = 'https://www.myntra.com/shoes'
driver.get(url)

time.sleep(2)
driver.find_element_by_xpath('//label[@class="common-customCheckbox"]/span[@data-colorhex="black"]').click() #selecting black color

time.sleep(2)
driver.find_element_by_xpath('//ul[@class="price-list"]/li[2]/label/div').click() #filter 6649-13099 price

time.sleep(4)

Brand=[]
Description=[]
Price=[]
for i in range(0,2):
    brand_name=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h3") #brand name
    for i in brand_name:
        Brand.append(i.text)
    desc=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/h4[1]") #description of shoes
    for i in desc:
        Description.append(i.text)
    Shoes_price=driver.find_elements_by_xpath("//div[@class='product-productMetaInfo']/div/span[1]") #shoes price
    for i in Shoes_price:
        a = i.text.split(" ") #splitting by space
        b = a[1][-3:]
        if(b.isnumeric()): #check if it's numeric
            pass
        else:
            a[1] = a[1][:-3]
        Price.append(a[1])
    
    nxt=driver.find_elements_by_xpath("//a[@rel='next']") #next button
    url=[]
    for i in nxt:
        url=i.get_attribute('href')
    time.sleep(2)
    driver.get(url)
    
Shoes=pd.DataFrame({})
Shoes['Brand'] = Brand
Shoes['Description'] = Description
Shoes['Price'] = Price
Shoes
```
